```python
import pandas as pd
import numpy as np
import csv
import seaborn
import matplotlib.pyplot as plt
```

```python
cps_df = pd.read_csv("cleaned_data/cps_cleaned.csv", names=["Row", "Demographic", 
                                                            "Total", "Universe", 
                                                            "Federal Government", "State Government", 
                                                            "Local Government", 
                                                            "Private (incl. Incorporated self employed)", 
                                                            "Self-Employed, (unincorporated)", 
                                                            "Without Pay", "Year", 
                                                            "Month", "Borough", "Sector"])
```

```python
cps_df.head(5)
```

```python
month_conv = {"jan" : 1,
              "feb" : 2,
              "march":3,
              "april":4,
              "may" : 5,
              "june": 6,
              "jul" : 7,
              "aug" : 8,
              "sep" : 9,
              "oct" : 10,
              "nov" : 11,
              "dec" : 12 
             }

cps_df["Month"] =  cps_df["Month"].apply(lambda x : month_conv[x])
cps_df["Year"] =  cps_df["Year"].apply(lambda x : x + 2000)
```

```python
cps_df['DATE'] = pd.to_datetime(cps_df[['Year', 'Month']].assign(DAY=1))
```

```python
cps_df["Month"].value_counts()
```

```python
cps_df.drop(columns = ["Year", "Month", "Without Pay"], inplace = True)
```
